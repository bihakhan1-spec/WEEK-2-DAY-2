### loop

```python
marks = {"Math": 90, "Physics": 75, "Chemistry": 88, "Biology": 95}
# Loop over keys (default)
for subjects in marks:
    print(subjects)
```

```python
marks = {"Math": 90, "Physics": 75, "Chemistry": 88, "Biology": 95}
# .items() -> key aur value dono ek saath (most useful!)
for subject, score in marks.items():
    print(subject, "->", score)
```

### Sets & Duplicates

```python
votes = ["PTI", "PMLN", "PTI", "PPP", "PMLN", "PTI"]
temp = set(votes)
print(temp)
print(type(temp))
```

```python
# Loop a set
for n in {'PMLN', 'PPP', 'PTI'}:
    print(n)
```

### Set Operations

```python
group_a = {"Ali", "Sara", "Bilal", "Hina"}
group_b = {"Bilal", "Hina", "Omar", "Zara"}

print(group_a | group_b)   # union -> 
print(group_a & group_b)   # intersection -> 




```

### list of dictionaries

```python
products = [
    {"name": "Pen",      "price": 50},
    {"name": "Notebook", "price": 120},
    {"name": "Bag",      "price": 800},
]
for p in products:
    print(p["name"], "-", p["price"])
```
