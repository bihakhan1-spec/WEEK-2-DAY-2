```python
Grocery = [
    {"name":"Meat", "price": 1000},
    {"name": "Veggies", "price": 500},
    {"name": "Cleaners", "price": 700},
    {"name": "Sauces", "price": 900},
]
# Append a new item
Grocery.append({"name": "Eggs", "price": 200})

# Update one value in one item
Grocery[0]["price"] = 2000

# Loop and print again
for item in Grocery:
    print(item["name"], "|", item["price"])
```

```python
# TODO: collect one field's values into a list using a loop
values = []
for row in Grocery:
     values.append(row["name"])   # apni key use karo

# TODO: print unique values with set()
print("Unique names: " , set(values))
```
